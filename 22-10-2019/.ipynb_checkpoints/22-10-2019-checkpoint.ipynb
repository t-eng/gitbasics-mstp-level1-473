```python
## line plot
import matplotlib.pyplot as m
x=[1,2,3,4,5]
y=[2,3,4,5,6]
m.plot(x,y)
m.show()
```

```python
## By using range
import matplotlib.pyplot as p
x=list(range(1,12,2))  ## 
y=list(range(20,32,2))
p.plot(x,y,'ys')
p.show()
```

```python
x=[i for i in range(1,11)]
print(x)
l=[i**0.5 for i in range(1,11)]
print(l)
p.plot(x,l,'go')
p.grid()
```

```python
## Label names and title for graph
import matplotlib.pyplot as p
x=[2,6]
y=[10,13]
p.plot(x,y)
p.xlabel('sqrt numbers')
p.ylabel('numbers')
p.title('squqre root of given numbers')
p.grid()
```

### Bar Graph

```python
import matplotlib.pyplot as p_l
x=list(range(1,20,3))
y=list(range(21,40,3))
p_l.bar(x,y)
p_l.show()
```

```python
## Scatter
x=list(range(1,20,2))
y=list(range(21,40,2))
p_l.scatter(x,y)
p_l.title('scatter graph')
```

### Histogram

```python
x=list(range(1,10,2))
p_l.hist(x)
p_l.title('Histogram')
```

```python
## plot for time and distance --->distance=2*time
time=[i  for i in range(10,22,2)]
print(time)
distance=[i*2 for i in range(10,22,2)]
print(distance)
p_l.plot(time,distance)
p_l.xlabel('Time')
p_l.ylabel('distance')
p_l.title('distance vs time')
```

### Getting images by using matplotlib

```python
import matplotlib.pyplot as pl
import matplotlib.image as k
pic=k.imread("download.jpg")
pl.imshow(pic)
```

```python
# Apllying psudo color to an image
import matplotlib.pyplot as pl
import matplotlib.image as k
x=k.imread('download.jpg')
c=x[:,:,0]
pl.imshow(c)

```

```python
import matplotlib.pyplot as pl
import matplotlib.image as l
x=l.imread('download.jpg')
c=x[:,:,0]
pl.imshow(c,cmap='hot')
```

```python
import matplotlib.pyplot as pl
import matplotlib.image as l
x=l.imread('download.jpg')
c=x[:,:,0]
pl.imshow(c,cmap='hot')
pl.colorbar()
```

#### matplotlib ticks ---->ticks means values within quardinates

```python
import matplotlib.pyplot as pl
x=list(range(5,50,5))
print(x)
y=list(range(1,10))
print(y)
pl.plot(y,x)
pl.title('example to ticks')
pl.xlabel('numbers')
pl.ylabel('values')
pl.show()
```

```python
import matplotlib.pyplot as pl
import numpy as np
x=str(pl.xticks(np.arange(1,11)))
y=(pl.yticks(np.arange(5,51,5)))
pl.title('values using numpy')
pl.xlabel('xticks')
pl.ylabel('yticks')
pl.plot('x','y')
pl.show()
```
