### Bar Graph

```python
import matplotlib.pyplot as p_l
x=list(range(1,20,3))
y=list(range(21,40,3))
p_l.bar(x,y)
p_l.show()
```

```python
## Scatter
x=list(range(1,20,2))
y=list(range(21,40,2))
p_l.scatter(x,y)
p_l.title('scatter graph')
```

### Histogram

```python
x=list(range(1,10,2))
p_l.hist(x)
p_l.title('Histogram')
```

```python
## plot for time and distance --->distance=2*time
time=[i  for i in range(10,22,2)]
print(time)
distance=[i*2 for i in range(10,22,2)]
print(distance)
p_l.plot(time,distance)
p_l.xlabel('Time')
p_l.ylabel('distance')
p_l.title('distance vs time')
```

#### matplotlib ticks ---->ticks means values within quardinates

```python
import matplotlib.pyplot as pl
x=[5,10,15,20,25,30,35,40,45,50]
print(x)
y=[9,8,7,6,5,4,3,4,6,4]
print(y)
pl.plot(y,x)
pl.title('example to ticks')
pl.xlabel('numbers')
pl.ylabel('values')
pl.show()
```

```python
import matplotlib.pyplot as pl
import numpy as np
pl.xticks(np.arange(10,20,1))
pl.yticks(np.arange(5,51,5))
pl.title('values using numpy')
pl.xlabel('xticks')
pl.ylabel('yticks')
pl.plot(y,x)
pl.show()
```

### Representing a sine wave
a=sin(wt):
a=amplitude:
w=frequency:
t=time:

```python
import numpy as np
import matplotlib.pyplot as pl
t=np.arange(1,10,0.1)
a=np.sin(t)
pl.plot(t,a)
pl.xlabel("time")
pl.ylabel('amplitude')
pl.title('amplitude vs time')
pl.grid()
```

```python
import pandas as pd
df=pd.read_csv("income.csv")
df
```

```python
import matplotlib.pyplot as pl
years=df.columns[2::]
allincome=df.iloc[1,2:]
pl.plot(years,allincome)
pl.xlabel('years')
pl.ylabel('income')
pl.title('income of Alaska from 2005 to 2013')
pl.savefig("income of Alaska from 2005 to 2013.jpg")
pl.show()
```

```python
import matplotlib.pyplot as pl
years=df.columns[-5:]
income=df.iloc[0:,2:]
pl.plot(years,income)
pl.xlabel('years')
pl.ylabel('income')
pl.title('income of all years from 2005 to 2013')
pl.savefig("income of all years from 2005 to 2013.jpg")
pl.show()
```
