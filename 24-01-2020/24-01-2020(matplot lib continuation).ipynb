```python
import numpy as np
t=np.arange(0.0,0.5,0.1)
s=np.sin(2*np.pi*t)
plt.subplot(2,2,3)
plt.plot(t,s,'rx',label='1st graph with sine wave')
plt.xlabel("t")
plt.ylabel("s")
plt.title("graph t vs s")
plt.legend()
t[0]=np.nan
t[1]=np.nan
plt.subplot(2,1,2)
plt.plot(t,s,label="2nd graph with sine wave")
plt.xlabel("t")
plt.ylabel("s")
plt.legend()
plt.show()
```

```python
x=[1,2,3,4]
y=[2,44,55,6]
z=[2,5,4,3]
fig=plt.figure()
ax=fig.add_subplot(212,projection='3d')
ax.plot(x,y,z)
plt.show()
plt.xlabel("x")
plt.ylabel("y")
#plt.zlabel("z")
plt.title("3d projection")
plt.show()
```

```python
x=[1,2,3,4]
y=[2,44,55,6]
z=[2,5,4,3]
dx=np.ones(4)
dy=np.zeros(4)
dz=np.ones(4)
fig=plt.figure()
ax=fig.add_subplot(212,projection='3d')
ax.plot(x,y,z,dx,dy,dz,color='m')
plt.show()
plt.xlabel("x")
plt.ylabel("y")
#plt.zlabel("z")
plt.title("3d projection")
plt.show()
```
